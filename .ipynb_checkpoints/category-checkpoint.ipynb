```python
import os 
import numpy as np
```

```python
#user_liking
#user_category
#photo_category
#YT_category --> identified with posted most 
categories = os.listdir('../Data/Image/ImageALL')
```

```python
photoSum = 0
for category in categories:
    photoSum += len(os.listdir('../Data/Image/ImageALL/'+category))
photoSum
```

```python
photo_category = np.zeros((photoSum,17))
ImgLists = []
i = 0 
for category in categories:
    ImgList = os.listdir('../Data/Image/ImageALL/'+category)
    ImgLists += ImgList
    if ',' in category: #複數的category 
        multi_category = [int(single)-1 for single in category.split(',')]
        print(multi_category)
        for ci in range(len(ImgList)):
            for single in multi_category:
                photo_category[i+ci][single] = 1
        print(photo_category[i:i+ci+1])
        i += ci+1
    else:
        single = int(category)-1
        print(single)
        for ci in range(len(ImgList)):
            photo_category[i+ci][single] = 1
        print(photo_category[i:i+ci+1])
        i += ci+1     
```

```python
ImgLists = np.array(ImgLists)
np.save('../Data/npy/ImgLists',ImgLists)
```

```python
# YT_category
dictImg = {}
for i in range(len(ImgLists)):
    dictImg[ImgLists[i]] = photo_category[i]
```

```python
YT_category = np.zeros((88,17))
YTs = os.listdir('../Data/Image/ImageYts/')
for i in range(len(YTs)):
    Imgs = os.listdir('../Data/Image/ImageYts/'+YTs[i])
    for Img in Imgs:
        YT_category[i] += dictImg[Img]
```

```python
YT_category
```

```python
YT_Normcategory = np.zeros((88,17))
YT_NormSum_category = np.zeros((88,17))
for i in range(len(YT_category)):
    YT_Normcategory[i] = YT_category[i]/np.sum(YT_category[i])
    YT_NormSum_category[i] = YT_category[i]/np.max(YT_category[i])
```

```python
np.save('../Data/npy/YT_category',YT_category)
np.save('../Data/npy/YT_Normcategory',YT_Normcategory)
np.save('../Data/npy/YT_NormSum_category',YT_NormSum_category)
```

```python
for i in range(len(YT_Normcategory)):
    if np.max(YT_Normcategory[i]) < 0.4:
        print(i, ' ',np.max(YT_Normcategory[i]))
        print(YT_Normcategory[i])
```
