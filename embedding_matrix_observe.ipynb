```python
def get_Matrix(path):
    matrics = np.load(path)
    #print(matrics.files)
    User_latent_factor = matrics['U']
    print('User latent factor shape',User_latent_factor.shape)
    YouTuber_latent_factor = matrics['Y']
    print('YouTuber latent factor shape',YouTuber_latent_factor.shape)
    Auxilary_latent_factor = matrics['A']
    print('Auxilary latent factor shape',Auxilary_latent_factor.shape)
    Embedding = matrics['E']
    print('Embedding shape',Embedding.shape)
    Wu = matrics['Wu']
    print('Wu shape',Wu.shape)
    Wy = matrics['Wy']
    print('Wy shape',Wy.shape)
    Wa = matrics['Wa']
    print('Wa shape',Wa.shape)
    Wv = matrics['Wv']
    print('Wv shape',Wv.shape)
    Beta = matrics['B']
    print('Beta shape',Beta.shape)
    return User_latent_factor,YouTuber_latent_factor,Auxilary_latent_factor,Embedding,Wu,Wy,Wa,Wv,Beta
```

```python
Embedding_Path = '../Data/npy/0.01_0.01_0.001_0.01_Edims200_15D.npz'
User_latent_factor,YouTuber_latent_factor,Auxilary_latent_factor,Embedding,Wu,Wy,Wa,Wv,Beta = get_Matrix(Embedding_Path)
```

## For Video

```python
#video_E* video_F
video_result = np.transpose(np.dot(video_E,np.transpose(video_F)))
videoList = []
for i in range(video_result.shape[0]):
    #print(len(video_result[i]))
    sorted_i = list(reversed(sorted(range(len(video_result[i])), key = lambda k : video_result[i][k])))
    print(sorted_i)
    #print([video_result[i][k] for k in sorted_i])
    print('--------------------------------------')
    videoList.append(sorted_i)
vdf = pd.DataFrame({'Embedding':videoList,'YouTuber':ytList})
vdf.to_csv('../Data/observe_Embedding/seperate/video.csv',index=False)
```
